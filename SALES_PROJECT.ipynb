<font color='blue' size=5><b>IMPORTING LIBRARIES </b> </font>

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns 
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
#drop blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#check for null values if there is any null value present
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Single/married'})
```

```python
# describe() (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

<font color='blue' size=5><b>DATA ANALYSIS</b> </font>

<font color='#1E90FF' size=5><b>1)GENDER</b> </font>

# <font color='#4682B4' size=3><b>(A) plotting a bar chart for Gender and it's count (SALE GENERETED BY M AND F) </b> </font>

```python
# plotting a bar chart for Gender and it's count

# Assuming df contains your data
ax = sns.countplot(x='Gender', data=df, palette=['skyblue', 'salmon'], edgecolor='black', linewidth=1.5, alpha=0.7)

# Adding labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars)

# Customizing ticks and labels
ax.tick_params(axis='x', labelsize=12)  # Adjust x-axis tick label size
ax.tick_params(axis='y', labelsize=12)  # Adjust y-axis tick label size

# Adding a title
plt.title('Gender Distribution', fontsize=16)

# Adding labels to axes
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adding gridlines
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Adding shadow to bars
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_color('gray')
    spine.set_linewidth(0.5)

# Adjusting plot layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by Gender and calculate the total amount for each gender
gender_total_amount = df.groupby('Gender')['Amount'].sum().reset_index()
```

```python
# Plotting the bar chart

ax = sns.barplot(x='Gender', y='Amount', data=gender_total_amount, palette=['skyblue', 'salmon'])
plt.show()
```

# <font color='#BLACK' size=5><b>*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*</b> </font>

<font color='#1E90FF' size=6><b>2)AGE</b> </font>

# <font color='#4682B4' size=3><b>(A) Number of Males and Females in Each Age Group </b> </font>

```python
gender_counts = df.groupby(['Age Group', 'Gender']).size().reset_index(name='Count')

# Plotting the countplot
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=gender_counts, x='Age Group', y='Count', hue='Gender', palette=['skyblue', 'salmon'])

# Adding labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars, label_type='center')

# Adding a title
plt.title('Number of Males and Females in Each Age Group', fontsize=16)

# Adding labels to axes
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Customizing legend
plt.legend(title='Gender', fontsize=12)

# Adjusting plot layout
plt.tight_layout()

# Show the plot
plt.show()
```

# <font color='#4682B4' size=3><b>(B) TOTAL AMOUNT GENERETED BY AGE GROUP</b> </font>

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Define a color palette with 5 colors
colors = sns.color_palette('Set2', n_colors=5)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, palette=colors)

# Adding labels on top of bars
for bars in ax.patches:
    ax.annotate(format(bars.get_height(), '.2f'), 
                (bars.get_x() + bars.get_width() / 2., bars.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

# Adding a title
plt.title('Total Amount by Age Group', fontsize=16)

# Adding labels to axes
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Total Amount', fontsize=14)

# Show the plot
plt.show()
```

# <font color='#BLACK' size=5><b>From above graphs we can see that most of the buyers are of age group between 26-35 yrs female</b> </font>

<font color='#1E90FF' size=6><b>3)STATES</b> </font>

# <font color='#4682B4' size=3><b>(A) Total number of orders from top 10 states</b> </font>

```python
# total number of orders from top 10 states
# Grouping by state and calculating the total number of orders
state_orders = df.groupby('State')['Orders'].sum().reset_index()

# Sorting the states by the number of orders in descending order and selecting the top 10
top_10_states = state_orders.sort_values(by='Orders', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='State', y='Orders', data=top_10_states, palette='viridis')

# Adding labels on top of bars
for bars in ax.patches:
    ax.annotate(format(bars.get_height(), '.0f'), 
                 (bars.get_x() + bars.get_width() / 2., bars.get_height()), 
                 ha='center', va='center', 
                 xytext=(0, 10), 
                 textcoords='offset points')

# Adding a title
plt.title('Total Number of Orders by State (Top 10)', fontsize=16)

# Adding labels to axes
plt.xlabel('State', fontsize=14)
plt.ylabel('Total Number of Orders', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

# <font color='#4682B4' size=3><b>(B) Total sales from the top 10 states</b> </font>

```python
# Grouping by state and calculating the total sales
state_sales = df.groupby('State')['Amount'].sum().reset_index()

# Sorting the states by total sales in descending order and selecting the top 10
top_10_states_sales = state_sales.sort_values(by='Amount', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_states_sales, x='State', y='Amount', palette='viridis')

# Adding labels and title
plt.xlabel('State', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.title('Total Sales from Top 10 States', fontsize=16)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

 <font color='#BLACK' size=5><b>*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
</b> </font>

<font color='#1E90FF' size=6><b>4)Marital Status</b> </font>

# <font color='#4682B4' size=3><b>(A) Marital Status Distribution: Graphs represent different marital status categories, such as "0," "1,"  etc. You can observe the count of individuals belonging to each category.</b> </font>

```python
#Marital Status


# Setting the color palette
palette = ['skyblue', 'salmon']

# Plotting the countplot
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Marital_Status', palette=palette)

# Adding labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars)

# Adding a title
plt.title('Count of Marital Status', fontsize=16)

# Adding labels to axes
plt.xlabel('Marital Status', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adjusting plot layout
plt.tight_layout()

# Show the plot
plt.show()
```

<font color='#4682B4' size=3><b>(B) Grouping by 'Marital_Status' and 'Gender' and getting the sum of 'Amount'</b> </font>

```python
# Grouping by 'Marital_Status' and 'Gender' and getting the sum of 'Amount'
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Define the color palette
palette = ['skyblue', 'salmon']

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender', palette=palette)

# Adding a title
plt.title('Total Amount by Marital Status and Gender', fontsize=16)

# Adding labels to axes
plt.xlabel('Marital Status', fontsize=14)
plt.ylabel('Total Amount', fontsize=14)

# Adjusting plot layout
plt.tight_layout()

# Show the plot
plt.show()
```

 <font color='#BLACK' size=5><b>From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
</b> </font>

<font color='#1E90FF' size=6><b>4)OCCUPATION</b> </font>

<font color='#4682B4' size=3><b>(A) The bars represent different occupation categories, such as "HEALTHCARE ," "GOVT," "AUTOMOBILE" etc. You can observe the count  of individuals belonging to each occupation category.</b> </font>

```python
#OCCUPATION

# Set the figure size
plt.figure(figsize=(20, 5))

# Plot the countplot
ax = sns.countplot(data=df, x='Occupation')

# Add labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars)

# Set the title
plt.title('Count of Individuals by Occupation', fontsize=16)

# Set labels for axes
plt.xlabel('Occupation', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show the plot
plt.show()



```

<font color='#4682B4' size=3><b>(B) Grouping data by 'Occupation' and calculating the sum of 'Amount' for each occupation</b> </font>

```python

# Grouping data by 'Occupation' and calculating the sum of 'Amount' for each occupation
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
# Adding a title
plt.title('Total Amount by Occupation', fontsize=16)

# Adding labels to axes
plt.xlabel('Occupation', fontsize=14)
plt.ylabel('Total Amount', fontsize=14)

# Show the plot
plt.show()
```

 <font color='#BLACK' size=5><b>From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
</b> </font>

<font color='#1E90FF' size=6><b>4)PRODUCT CATEGORY</b> </font>

<font color='#4682B4' size=3><b>(B) Chart displaying the distribution of products across different product categories</b> </font>

```python
##PRODUCT CATEGORY

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)


# Set plot title
plt.title('Count of Products by Product Category', fontsize=16)

# Set labels for axes
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Count', fontsize=14)


# Rotate x-axis labels
plt.xticks(rotation=50)

# Show the plot
plt.show()

```

<font color='#4682B4' size=3><b>(B) Chart showing the top 10 product categories by total sales amount</b> </font>

```python
# Grouping by 'Product_Category' and getting the sum of 'Amount', then sorting
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set the figure size
sns.set(rc={'figure.figsize':(20, 5)})

# Plot the bar plot
sns.barplot(data=sales_state, x='Product_Category', y='Amount', palette='Set2')

# Set plot title
plt.title('Top 10 Product Categories by Sales Amount', fontsize=16)

# Set labels for axes
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)

# Show the plot
plt.show()
```

```python
#Top 10 Most Sold Products (same as above charts)


# Set the figure size
plt.figure(figsize=(12, 7))

# Plot the bar plot for top 10 most sold products
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar', color='skyblue')

# Set plot title
plt.title('Top 10 Most Sold Products', fontsize=16)

# Set labels for axes
plt.xlabel('Product ID', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)

# Show the plot
plt.show()
```
